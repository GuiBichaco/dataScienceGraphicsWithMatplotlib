#Questions:







```python
#1: Qual é a distribuição de gêneros das séries na Netflix?
genre_counts = df['genre'].value_counts()
plt.figure(figsize=(10, 6))
genre_counts.plot(kind='bar', color='skyblue')
plt.title('Distribuição de Gêneros das Séries na Netflix')
plt.xlabel('Gênero')
plt.ylabel('Número de Séries')
plt.xticks(rotation=45)
plt.show()
```

```python
#2: Qual é a distribuição de idiomas das séries na Netflix?
language_counts = df['language'].value_counts()
plt.figure(figsize=(8, 5))
language_counts.plot(kind='pie', autopct='%1.1f%%', colors=['lightcoral', 'lightblue', 'lightgreen'])
plt.title('Distribuição de Idiomas das Séries na Netflix')
plt.ylabel('')
plt.show()
```

```python
#3: Como a pontuação no IMDb varia ao longo dos anos?
plt.figure(figsize=(10, 6))
plt.scatter(df['year'], df['imdb_score'], color='orange')
plt.title('Pontuação no IMDb ao Longo dos Anos')
plt.xlabel('Ano de Lançamento')
plt.ylabel('Pontuação IMDb')
plt.grid(True)
plt.show()
```

```python
#4: Qual é a relação entre a duração e a pontuação no IMDb das séries?
plt.figure(figsize=(10, 6))
plt.scatter(df['runtime'], df['imdb_score'], color='green')
plt.title('Relação entre Duração e Pontuação no IMDb das Séries')
plt.xlabel('Duração (minutos)')
plt.ylabel('Pontuação IMDb')
plt.grid(True)
plt.show()
```

```python
#5: Como as séries estão distribuídas ao longo dos anos de estreia?
year_counts = df['year'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
year_counts.plot(kind='line', marker='o', color='purple')
plt.title('Distribuição das Séries ao Longo dos Anos de Estreia')
plt.xlabel('Ano de Estreia')
plt.ylabel('Número de Séries')
plt.grid(True)
plt.show()
```

```python
#6: Qual é a distribuição das pontuações no IMDb das séries?
plt.figure(figsize=(10, 6))
plt.hist(df['imdb_score'], bins=10, color='salmon', edgecolor='black')
plt.title('Distribuição das Pontuações no IMDb das Séries')
plt.xlabel('Pontuação IMDb')
plt.ylabel('Frequência')
plt.grid(True)
plt.show()
```

```python
#7: Qual é a média da duração das séries por gênero?
genre_runtime_mean = df.groupby('genre')['runtime'].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
genre_runtime_mean.plot(kind='barh', color='gold')
plt.title('Média da Duração das Séries por Gênero')
plt.xlabel('Duração Média (minutos)')
plt.ylabel('Gênero')
plt.grid(True)
plt.show()
```
